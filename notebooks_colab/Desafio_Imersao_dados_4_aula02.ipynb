### 1. Criar uma escala de valores em milhões

```python
dados_vendas.sample(4)
```

```python
d_vendas = dados_vendas.drop(labels=["Valor_anuncio", "Tipo_anuncio", "Valor"], axis=1)
d_vendas.rename(columns={'Valor_anuncio_float': 'Valor'}, inplace=True)

escala_milhoes = (d_vendas['Valor']/1000000)
d_vendas['Valor_milhoes'] = escala_milhoes
d_vendas.sample(4)
```

### 2. Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

```python
plt.figure(figsize=(10,10))
sns.set(style='whitegrid')
ax = sns.histplot(data=d_vendas, x="Valor_milhoes", color="#02779e", alpha=0.6, linewidth=0.5, binwidth=0.2, kde=True)
ax.set_title("Quantidade de imóveis por faixa de preço")

plt.xlim((-0.05, 10))
plt.xlabel("Valor dos imóveis (em milhões de reais)")
plt.ylabel("Quantidade de imóveis")
plt.ticklabel_format(style='plain')
sns.despine(left=True, bottom=True)
plt.show()
```

### 3. Preço do metro quadrado por bairro e plotar em um gráfico ideal

### 4. Explorar as bibliotecas de visualizações e colocar as suas conclusão

```python
d_vendas["Valor_m2"] = (d_vendas['Valor']/d_vendas['Metragem'])
valor_m2_medio_por_bairros = d_vendas[["Bairro", "Valor_m2"]].groupby("Bairro", as_index=False).mean()
valor_m2_medio_por_bairros.rename(columns={'Valor_m2': 'Valor_medio_m2'}, inplace=True)
valor_m2_medio_por_bairros.sample(6)
```

```python
d_vendas['Bairro'].value_counts()

contagem_imoveis_por_bairro = d_vendas.groupby("Bairro", as_index=False).size().sort_values(by='size', ascending=False)
contagem_imoveis_por_bairro.rename(columns={'size': 'Qtd_imoveis'}, inplace=True)
top_12_bairros_qtd = contagem_imoveis_por_bairro.head(12)

merge = pd.merge(top_12_bairros_qtd, valor_m2_medio_por_bairros, how='inner', on='Bairro')
merge["Valor_medio_m2_por_imovel"] = (merge['Valor_medio_m2']/merge['Qtd_imoveis'])
merge
```

```python
ord_merge = merge.sort_values('Valor_medio_m2', ascending = False)
sns.set(style='whitegrid')
plt.figure(figsize=(10,10))
ax = sns.barplot(data=ord_merge, x="Valor_medio_m2", y="Bairro", color="#028cba")
ax.set(ylabel="", xlabel="Valor médio do m² em reais")
plt.title("1. Preço médio do m² nos bairros com mais imóveis à venda", fontsize=14)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set(style='whitegrid')

ax = sns.scatterplot(x="Valor_medio_m2", y="Qtd_imoveis", hue="Bairro", size="Qtd_imoveis", sizes=(40, 400), palette="muted", data=ord_merge)

ax.set(ylabel="Quantidade de imóveis", xlabel="Valor médio do m² em reais")
plt.title("2. Preço médio do m² nos bairros com mais imóveis à venda", fontsize=14)
sns.despine(left=True, bottom=True) 
plt.show()
```

```python
ord_merge_q = merge.sort_values('Valor_medio_m2_por_imovel', ascending = False)
plt.figure(figsize=(10,10))
ax = sns.barplot(data=ord_merge_q, x="Valor_medio_m2_por_imovel", y="Bairro", color="#028cba")
ax.set(ylabel="", xlabel="Valor médio do m² por imóvel (em reais)")
plt.title("3. Preço médio do m² nos bairros com mais imóveis à venda considerando-se a quantidade de imóveis", fontsize=14)
sns.despine(left=True, bottom=True)
plt.show()
```

Para melhorar a visualização, fora retirada uma amostra que compreende os bairros com maior quantidade de imóveis à venda. A partir dessa amostra, foram selecionados os valores médios do metro quadrado de cada um desses 12 bairros.

Ao dispor as informações ranqueadas por valor do m² (gráfico 01), vê-se que há uma discrepância do provável real valor médio, haja vista a diferença na quantidade de imóveis em cada bairro, mesmo na amostra.

Ao relacionar a quantidade de imóveis e os valores (gráfico 2), pode-se ter uma melhor noção da distribuição dos valores/ quantidade, no entanto, a exibição dos valores no gráfico pode ser um pouco confusa devido ao número de bairros observados.

Possivelmente, o que pode representar uam maior fidedignidade na distibuição dos valores é o que é apresentado no gráfico 03, onde os valores médios do metro quadrado é dividido pela quantidade de imóveis em cada localidade.
